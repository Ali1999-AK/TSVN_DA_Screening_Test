```python
def convert(x):
    if "/" in x:
        v ="20" + "-".join(x.split("/")[::-1])
        return v
    elif ":" in x:
        v = x.split(":")[0][:10]
        return v
    else:
        return x[:4]+"-"+x[4:6]+"-"+x[6:8]
```

```python
data['join_date'] = data['join_date'].apply(convert)
```

```python
def title(x):
    return x.title()
```

```python
data['full_name'] = data['full_name'].apply(title)
```

```python
def email(x):
    v = x.split(" ")
    v.append('@yourcompany.com')
    return "".join(v).lower()
```
